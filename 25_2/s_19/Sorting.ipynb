```
import random
import time

# ----------- 1. Bubble Sort -----------
def bubble_sort(arr):
    n = len(arr)
    count = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return count

# ----------- 2. Quick Sort -----------
def quick_sort(arr):
    count = [0]

    def _quick_sort(a, low, high):
        if low < high:
            p = partition(a, low, high)
            _quick_sort(a, low, p - 1)
            _quick_sort(a, p + 1, high)

    def partition(a, low, high):
        pivot = a[high]
        i = low - 1
        for j in range(low, high):
            count[0] += 1
            if a[j] <= pivot:
                i += 1
                a[i], a[j] = a[j], a[i]
        a[i + 1], a[high] = a[high], a[i + 1]
        return i + 1

    _quick_sort(arr, 0, len(arr) - 1)
    return count[0]

# ----------- 3. Merge Sort -----------
def merge_sort(arr):
    count = [0]

    def _merge_sort(a):
        if len(a) > 1:
            mid = len(a) // 2
            L = a[:mid]
            R = a[mid:]

            _merge_sort(L)
            _merge_sort(R)

            i = j = k = 0
            while i < len(L) and j < len(R):
                count[0] += 1
                if L[i] < R[j]:
                    a[k] = L[i]
                    i += 1
                else:
                    a[k] = R[j]
                    j += 1
                k += 1
            while i < len(L):
                a[k] = L[i]
                i += 1
                k += 1
            while j < len(R):
                a[k] = R[j]
                j += 1
                k += 1

    _merge_sort(arr)
    return count[0]

# ----------- 4. Heap Sort -----------
def heapify(arr, n, i, count):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and arr[l] > arr[largest]:
        largest = l
    if r < n and arr[r] > arr[largest]:
        largest = r
    if largest != i:
        count[0] += 1
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest, count)

def heap_sort(arr):
    n = len(arr)
    count = [0]
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i, count)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0, count)
    return count[0]

# ----------- Benchmarking -----------
def benchmark():
    algorithms = {
        "Bubble Sort": bubble_sort,
        "Quick Sort": quick_sort,
        "Merge Sort": merge_sort,
        "Heap Sort": heap_sort
    }

    size = 1000
    results = []

    for name, func in algorithms.items():
        data = [random.randint(0, 10000) for _ in range(size)]
        start = time.time()
        iterations = func(data.copy())
        elapsed = time.time() - start
        results.append((name, round(elapsed, 5), iterations))

    print("\nBenchmark results:")
    print("Algorithm\tTime (s)\tIterations")
    for name, t, it in results:
        print(f"{name:12s}\t{t:.5f}\t{it}")

if __name__ == "__main__":
    benchmark()
```
